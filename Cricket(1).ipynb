```python
import pandas as pd

# Example data
data = india['Teams']

# Create dataframe
df = pd.DataFrame(data, columns=['Teams'])

# Function to count teams
def count_teams(team_list):
    flattened_list = [team for sublist in team_list for team in sublist]
    return len(set(flattened_list))

# Calculate team count and add as a new column
df['team_count'] = df['Teams'].apply(count_teams)
```

**INNINGS**

```python
import pandas as pd

# Assuming you have loaded your data into a DataFrame called df

# List of columns containing innings values
innings_columns = ['Inns.0_x', 'Inns.1_x', 'Inns.2_x', 'Inns.3_x', 'Inns.4_x', 'Inns.5_x',
                   'Inns.0_y', 'Inns.1_y', 'Inns.2_y', 'Inns.3_y', 'Inns.4_y', 'Inns.5_y']

# Replace non-numeric values or missing values with 0
india[innings_columns] = india[innings_columns].apply(pd.to_numeric, errors='coerce').fillna(0)

# Create the 'innings' column by summing all innings values
india['innings'] = india[innings_columns].sum(axis=1)
```

```python
# List of columns to drop
columns_to_drop = ['Inns.0_x', 'Inns.1_x', 'Inns.2_x', 'Inns.3_x', 'Inns.4_x', 'Inns.5_x',
                   'Inns.0_y', 'Inns.1_y', 'Inns.2_y', 'Inns.3_y', 'Inns.4_y', 'Inns.5_y']

# Drop the columns
india = india.drop(columns=columns_to_drop)

```

**RUNS**

```python
import pandas as pd

# Assuming you have loaded your data into a DataFrame called india

# List of columns containing runs values
runs_columns = ['Runs.0_x', 'Runs.1_x', 'Runs.2_x', 'Runs.3_x', 'Runs.4_x', 'Runs.5_x',
                'Runs.0_y', 'Runs.1_y', 'Runs.2_y', 'Runs.3_y', 'Runs.4_y', 'Runs.5_y']

# Replace non-numeric values or missing values with 0
india[runs_columns] = india[runs_columns].apply(pd.to_numeric, errors='coerce').fillna(0)

# Create the 'runs' column by summing all runs values
india['runs'] = india[runs_columns].sum(axis=1)
```

```python
# List of columns to drop
columns_to_drop = ['Runs.0_x', 'Runs.1_x', 'Runs.2_x', 'Runs.3_x', 'Runs.4_x', 'Runs.5_x',
                'Runs.0_y', 'Runs.1_y', 'Runs.2_y', 'Runs.3_y', 'Runs.4_y', 'Runs.5_y']

# Drop the columns
india = india.drop(columns=columns_to_drop)
```

  **WICKETS**

```python

wic_columns = ['Wkts.0', 'Wkts.1', 'Wkts.2', 'Wkts.3', 'Wkts.4', 'Wkts.5']

# Replace non-numeric values or missing values with 0
india[wic_columns] = india[wic_columns].apply(pd.to_numeric, errors='coerce').fillna(0)

# Create the 'wicket' column by summing all wickets values
india['wicket'] = india[wic_columns].sum(axis=1)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have loaded your data into a DataFrame called india

# Plotting the distribution of player age
plt.figure(figsize=(8, 6))
india['Age_in_years'].hist(bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Player Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()

# Plotting the distribution of batting style
plt.figure(figsize=(10, 6))
batting_style_counts = india['Batting Style'].value_counts()
batting_style_counts.plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title('Distribution of Batting Styles')
plt.xlabel('Batting Style')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.grid(False)
plt.tight_layout()
plt.show()

# Plotting the distribution of bowling style
plt.figure(figsize=(10, 6))
bowling_style_counts = india['Bowling Style'].value_counts()
bowling_style_counts.plot(kind='bar', color='lightcoral', edgecolor='black')
plt.title('Distribution of Bowling Styles')
plt.xlabel('Bowling Style')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.grid(False)
plt.tight_layout()
plt.show()
```

```python

india['runs'] = pd.to_numeric(india['runs'], errors='coerce')

# Plotting the distribution of runs
plt.figure(figsize=(10, 6))
india['runs'].hist(bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Runs')
plt.xlabel('Runs')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()


# Plotting the distribution of wickets
india['wicket'] = pd.to_numeric(india['wicket'], errors='coerce')
plt.figure(figsize=(10, 6))
india['wicket'].hist(bins=20, color='lightgreen', edgecolor='black')
plt.title('Distribution of Wickets')
plt.xlabel('Wickets')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()


# Summary statistics
print("Summary statistics for runs:")
print(india['runs'].describe())

print("\nSummary statistics for wickets:")
print(india['wicket'].describe())
```

```python

# Select relevant columns for analysis
selected_columns = ['Age_in_years', 'Batting Style', 'Bowling Style', 'wicket', 'runs']
data = india[selected_columns]

# Convert categorical variables to numerical using one-hot encoding
data = pd.get_dummies(data, columns=['Batting Style', 'Bowling Style'])

# Remove rows with missing values
data = data.dropna()

# Plot correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()

# Plot scatter plots for selected attributes
sns.pairplot(data, vars=['Age_in_years', 'runs', 'wicket'], kind='scatter', diag_kind='kde')
plt.suptitle('Pairplot of Age, Runs, and Wickets', y=1.02)
plt.show()
```
